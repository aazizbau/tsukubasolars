# MLP Performance Visualization

Hard-coded performance metrics for the MLP model (before cleaning):

```
version accuracy precision recall f1 tn fp fn tp
before  0.9541  0.9057    0.96 0.932 195 10 4 96
```

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Hard-coded metrics
metrics = {
    "version": "before",
    "accuracy": 0.9541,
    "precision": 0.9057,
    "recall": 0.96,
    "f1": 0.932,
    "tn": 195,
    "fp": 10,
    "fn": 4,
    "tp": 96,
}

# Bar plot for accuracy/precision/recall/f1
scores = pd.Series({
    "Accuracy": metrics["accuracy"],
    "Precision": metrics["precision"],
    "Recall": metrics["recall"],
    "F1": metrics["f1"],
})

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].set_title("MLP Performance Metrices")
scores.plot(kind="bar", ax=axes[0], color=[
            "#4c72b0", "#55a868", "#c44e52", "#8172b2"])
axes[0].set_ylim(0, 1)
axes[0].set_ylabel("Score")
for idx, val in enumerate(scores):
    axes[0].text(idx, val + 0.02, f"{val:.3f}", ha="center", va="bottom")

cm = np.array([[metrics["tn"], metrics["fp"]], [metrics["fn"], metrics["tp"]]])
axes[1].imshow(cm, cmap="Blues")
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(["Negative", "Positive"])
axes[1].set_yticks([0, 1])
axes[1].set_yticklabels(["Negative", "Positive"])
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")
axes[1].set_title("Confusion Matrix")
for i in range(2):
    for j in range(2):
        axes[1].text(j, i, cm[i, j], ha="center", va="center", color="black")
plt.tight_layout()
plt.show()
````
