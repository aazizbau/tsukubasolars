# Tsukuba City Solar Panels 2017

Visualize the 2017 RF solar panel classification with city boundary, north arrow, scale bar, and graticule.

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.lines import Line2D
import rasterio
import geopandas as gpd
from pathlib import Path

# Resolve repo root (handles running from notebooks/ or repo root)
candidates = [Path.cwd(), Path.cwd().parent, Path.cwd().parent.parent]
repo_root = None
for cand in candidates:
    if (cand / 'map').exists():
        repo_root = cand
        break
if repo_root is None:
    raise FileNotFoundError('Cannot locate repo root with a map/ directory')

raster_path = repo_root / 'map/tsukuba_solarpanel_class_rf_grid_2017.tif'
boundary_path = repo_root / 'map/tsukuba_gp.gpkg'

cmap = ListedColormap([[0, 0, 0, 0], [1.0, 0.55, 0.0, 0.9]])

with rasterio.open(raster_path) as src:
    data = src.read(1)
    bounds = src.bounds
    extent = [bounds.left, bounds.right, bounds.bottom, bounds.top]
    crs = src.crs

gdf = gpd.read_file(boundary_path) if boundary_path.exists() else None
if gdf is not None and crs is not None:
    gdf = gdf.to_crs(crs)

fig, ax = plt.subplots(figsize=(12, 14), dpi=300)
ax.set_title('Tsukuba City Solar Panels 2017', fontsize=20)
ax.imshow(data, cmap=cmap, extent=extent, origin='upper', interpolation='nearest')
if gdf is not None:
    gdf.plot(ax=ax, facecolor='none', edgecolor='firebrick', linewidth=1.0)

# Grids with labels
ax.set_xticks(ax.get_xticks())
ax.set_yticks(ax.get_yticks())
ax.grid(True, which='both', color='gray', linestyle=':', linewidth=0.5)
ax.tick_params(axis='both', labelsize=8, colors='gray')

# North arrow in upper right pointing up
ax.annotate(
    'N',
    xy=(0.9, 0.35),
    xytext=(0.9, 0.15),
    arrowprops=dict(facecolor='black', width=2, headwidth=8),
    ha='center',
    va='center',
    fontsize=12,
    xycoords=ax.transAxes,
)

# Scale bar (approx visual) at bottom center with three ticks
bar_y = 0.06
ax.add_line(Line2D([0.35, 0.65], [bar_y, bar_y], transform=ax.transAxes, color='black', linewidth=3))
for frac in [0.35, 0.45, 0.55, 0.65]:
    ax.add_line(Line2D([frac, frac], [bar_y - 0.01, bar_y + 0.01], transform=ax.transAxes, color='black', linewidth=2))
ax.text(0.5, bar_y + 0.02, '2 km', transform=ax.transAxes, ha='center', va='bottom', fontsize=10)

plt.show()
```
